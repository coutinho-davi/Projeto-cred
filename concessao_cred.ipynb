```python
import pandas as pd
import numpy as np
import joblib

model = joblib.load("Project_Credit/modelo_credito.pkl")
```

### Carregar dados do cliente

```python
novo_cliente = pd.DataFrame([{
    "person_age": 30,
    "person_income": 50000,
    "person_emp_length": 5,
    "loan_amnt": 10000,
    "loan_int_rate": 12.5,
    "loan_percent_income": 0.2,
    "cb_person_cred_hist_length": 10,
    "cb_person_default_on_file": 0
}])
```

### Classificação de risco + Predição

```python
# quanto maior o risco, menor a nota do empréstimo

def model_loangrade(risk):
    if risk < 3:
        return """A. 
O risco de crédito é minimo."""
    elif risk < 6:
        return """B. 
O risco de crédito é baixo."""              
    elif risk < 12:
        return """C. 
O risco de crédito é moderado."""
    elif risk < 25:
        return """D. 
O risco de crédito é alto."""
    elif risk < 50:
        return """E. 
O risco de crédito é muito alto."""
    elif risk < 75:
        return """F. 
O risco de crédito é extremamente alto."""
    else:
        return """G. 
O risco de crédito é impraticável."""

# Predição de risco para o novo cliente

risk = model.predict_proba(novo_cliente)[0][1] * 100
```

### Resultado

```python
print(f"Risco de inadimplência: {risk:.2f}%")
print(f"Nota de crédito: {model_loangrade(risk)}")
```
