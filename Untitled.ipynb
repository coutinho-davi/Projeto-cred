```python
import pandas as pd
import numpy as np
import xlrd as xlrd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
import os
import csv
```

```python
df_credito = pd.read_csv("Projeto_Credito/Credito/credit_risk_dataset.csv")
df_ipca = pd.read_excel("Projeto_Credito/Ipca/ipca_202602SerieHist.xls",skiprows=5)
df_selic = pd.read_csv( "Projeto_Credito/Selic/bcdata.sgs.11.csv", sep=";",  decimal="," ) 
```

Importando e Tratando IPCA

```python
df_ipca = df_ipca.dropna(how="all")
df_ipca = df_ipca.astype(str).apply(lambda x: x.str.strip())
```

```python
df_ipca["ano"] = df_ipca["Unnamed: 0"].where(
    df_ipca["Unnamed: 0"].str.match(r"^\d{4}$")
).ffill()
```

```python
df_ipca["mes"] = df_ipca["Unnamed: 1"]
 
meses_validos = [
    "JAN","FEV","MAR","ABR","MAI","JUN",
    "JUL","AGO","SET","OUT","NOV","DEZ"
]
```

```python
df_ipca = df_ipca[df_ipca["mes"].isin(meses_validos)]
```

```python
df_ipca = df_ipca[df_ipca["ano"].isin(["2024", "2025", "2026"])]
```

```python
df_ipca["ipca"] = (
    df_ipca[3]
    .str.replace(",", ".", regex=False)
    .astype(float)
)
```

```python
mapa_meses = {
    "JAN": 1, "FEV": 2, "MAR": 3, "ABR": 4,
    "MAI": 5, "JUN": 6, "JUL": 7, "AGO": 8,
    "SET": 9, "OUT": 10, "NOV": 11, "DEZ": 12
}
```

```python
df_ipca["mes_num"] = df_ipca["mes"].map(mapa_meses)
```

```python
df_ipca["data"] = pd.to_datetime(
    df_ipca["ano"] + "-" + df_ipca["mes_num"].astype(str),
    format="%Y-%m"
)
```

```python
df_ipca.head(14)
```

```python
df_ipca_final = (
    df_ipca[["data", "ano", "mes", "ipca"]]
    .sort_values("data")
    .reset_index(drop=True)
)
 
df_ipca_final
```

```python
#Renomear colunas IPCA:
 
df_ipca = df_ipca.rename(columns={
    "ano": "year",
    "mes": "month"
})
```

```python
mapa_meses = {
    "JAN": 1, "FEV": 2, "MAR": 3, "ABR": 4,
    "MAI": 5, "JUN": 6, "JUL": 7, "AGO": 8,
    "SET": 9, "OUT": 10, "NOV": 11, "DEZ": 12
}
 
df_ipca["month"] = df_ipca["month"].map(mapa_meses)
```

```python
#Garantir tipos coreetos
 
df_ipca["year"] = df_ipca["year"].astype(int)
df_ipca["month"] = df_ipca["month"].astype(int)
df_ipca["ipca"] = pd.to_numeric(df_ipca["ipca"], errors="coerce")
```

```python
df_ipca_final.to_csv( "Projeto_Credito/Ipca/ipca_2024_2026_tratado.csv", 
index=False, 
encoding="utf-8" ) 
```

Tratamento da Selic

```python
df_selic["data"] = pd.to_datetime(df_selic["data"], dayfirst=True)
 
df_selic["year"] = df_selic["data"].dt.year
df_selic["month"] = df_selic["data"].dt.month
```

```python
df_selic.head(5)
 
```

```python
df_selic_mensal = (
    df_selic
    .groupby(["year", "month"])["valor"]
    .mean()
    .reset_index()
    .rename(columns={"valor": "selic_media"})
)
```

```python
df_selic_mensal.head(5)
```

Crédito

```python
df_credito.columns
df_ipca.columns
df_selic_mensal.columns
```

```python
df_credito[["year", "month"]].head()
df_ipca[["year", "month", "ipca"]].head()
df_selic_mensal[["year", "month", "selic_media"]].head()
```

```python
# Colunas numéricas que podem vir com vírgula
colunas_decimal = [
    "loan_int_rate",
    "loan_percent_income",
    "person_emp_length"
]
 
for col in colunas_decimal:
    df_credito[col] = (
        df_credito[col]
        .astype(str)
        .str.replace(",", ".", regex=False) 
        .replace("nan", np.nan)
        .astype(float)
    )
```

Integração das bases (Join)

```python
#Crédito + IPCA
 
df_integrado = df_credito.merge(
    df_ipca,
    on=["year", "month"],
    how="left"
)
```

```python
#Adicionanddo Selic

df_integrado = df_integrado.merge(
    df_selic_mensal,
    on=["year", "month"],
    how="left"
)
```

Criação de DB

```python
caminho_db = os.path.join(pasta_data, "credit_risk.db")
print(f"DB será salvo em: {caminho_db}")
```

```python
#Criação e conexao com o banco de dados SQLite

conexao = sqlite3.connect(caminho_db)
cursor = conexao.cursor()
```

```python
cursor.execute("""
        CREATE TABLE IF NOT EXISTS credito_integrado (
            person_age INTEGER,
            person_income REAL,
            person_home_ownership TEXT,
            person_emp_length REAL,
            loan_intent TEXT,
            loan_grade TEXT,
            loan_amnt REAL,
            loan_int_rate REAL,
            loan_status INTEGER,
            loan_percent_income REAL,
            cb_person_default_on_file TEXT,
            cb_person_cred_hist_length INTEGER,
            date INTEGER,
            year INTEGER,
            month INTEGER,
            unnamed_0 TEXT,
            unnamed_1 TEXT,
            ipca REAL,
            no REAL,
            dez_93_100 REAL,
            no_1 REAL,
            n12 REAL,
            ipca_tipo TEXT,
            mes_num INTEGER,
            data_ref TEXT,
            selic_media REAL
        );
    """)

print("Tabela criada com sucesso!")
```

```python
with open("data/processed/credit_risk_integrado_v1.csv", newline="", encoding="utf-8") as file:
    reader = csv.reader(file, delimiter=";")
    header = next(reader)
 
    rows = [row[:26] for row in reader]
 
    cursor.executemany("""
        INSERT INTO credito_integrado (
            person_age,
            person_income,
            person_home_ownership,
            person_emp_length,
            loan_intent,
            loan_grade,
            loan_amnt,
            loan_int_rate,
            loan_status,
            loan_percent_income,
            cb_person_default_on_file,
            cb_person_cred_hist_length,
            date,
            year,
            month,
            unnamed_0,
            unnamed_1,
            ipca,
            no,
            dez_93_100,
            no_1,
            n12,
            ipca_tipo,
            mes_num,
            data_ref,
            selic_media
        ) VALUES (
            ?, ?, ?, ?, ?, ?, ?, ?, ?, ?,
            ?, ?, ?, ?, ?, ?, ?, ?, ?, ?,
            ?, ?, ?, ?, ?, ?
        )
    """, rows)
```

```python
conexao.commit()

print("✅ Tabelas criadas com sucesso!")
```
